```python
# GENERATING SINUSOID PITCHES WITH PLAYABLE AUDIO

import numpy as np
import IPython.display as ipd # Used to play audio directly inside a Jupyter Notebook

def generate_sinusoid_pitches(pitches=[69], dur=0.5, Fs=4000, amp=1):
    """Generation of sinusoids for a given list of MIDI pitches

    Notebook: C1/C1S1_MusicalNotesPitches.ipynb

    Args:
        pitches (list): List of MIDI pitches (Default value = [69])
        dur (float): Duration (in seconds) of each sinusoid (Default value = 0.5)
        Fs (scalar): Sampling rate (Default value = 4000)
        amp (float): Amplitude of generated signal (Default value = 1)

    Returns:
        x (np.ndarray): Signal
        t (np.ndarray): Time axis (in seconds)
    """
    N = int(dur * Fs) # number of samples=duration×sampling rate =====> discrete points to store the duration (right now: one second) of sound 
    t = np.arange(N) / Fs 
    x = []
    for p in pitches:
        freq = 2 ** ((p - 69) / 12) * 440
        x = np.append(x, np.sin(2 * np.pi * freq * t)) # actual sound signal
    x = amp * x / np.max(x)
    return x, t

dur = 1
Fs = 22050

pitches = [36,48,60,72,84,96,108]
x, t = generate_sinusoid_pitches(pitches=pitches, dur=dur, Fs=Fs, amp=0.5)
print('Pitch class C = {..., C1, C2, C3, C4, C5, C6, C7, ...}', flush=True)
ipd.display(ipd.Audio(data=x, rate=Fs))
```

```python
# Create sinusoid scales

dur = 0.5
Fs = 22050
x_maj, t = generate_sinusoid_pitches(pitches=[60,62,64,65,67,69,71,72], dur=dur, Fs=Fs, amp=0.5)
x_min, t = generate_sinusoid_pitches(pitches=[60,62,63,65,67,68,70,72], dur=dur, Fs=Fs, amp=0.5)

print('C major scale', flush=True)
ipd.display(ipd.Audio(data=x_maj, rate=Fs))
print('C minor scale', flush=True)
ipd.display(ipd.Audio(data=x_min, rate=Fs))
```

```python
# Sonification of chromatic scale, up- and downwards

dur = 0.125
Fs = 22050
up_pitches = list(range(128))
down_pitches = up_pitches[::-1]
pitches = up_pitches + down_pitches
x_chromatic, t = generate_sinusoid_pitches(pitches=pitches, dur=dur, Fs=Fs, amp=0.5)

print('Sinusoidal sonification of the chromatic scale + downwards:', flush=True)
ipd.display(ipd.Audio(data=x_chromatic, rate=Fs))
```

```python
# 1.9
import numpy as np
"""
Calculate differences (in cents) between the first 16 harmonics of C2 and the center frequencies of the closest notes of the 12-tone equal tempered scale.
What are the corresponding differences when considering the harmonics of another note such as b-flat4
"""
c2 = 65.4
harmonic_multipliers = range(1,17)
for i in harmonic_multipliers:
    harmonic = c2 * i
    print(round(harmonic, 2), "Hz")

# by solving the formula, we see that the differences are independent of the pitch p; it doesn't matter if we start with a c2 or b-flat4! Compare solution sheet
# implement matlab solution as py code

for m in range(1,17):
    diff = (np.log2(m) - 1/12) * 1200
    diff = diff % 100
    if diff > 50:
        diff = diff-100
    print(f"m = {m:2d}, diff = {diff:+6.2f}")    
```
